```python
# Task1

'''
Create a dataframe with the 10 names of pupils and the number of points 
they obtained in the exam. Then check what the mean and median of the 
results were.
'''

import pandas as pd

data = pd.DataFrame({'name': ['Janko', 'Martin', 'Janka', 'Tomas', 'Mirko', 'Peter', 'Lenka', 'Jarka',], 'points':[5,6,3,8,8,9,4,8]})

data.points.mean()
data.describe()
```

```python
# Task4

'''
Sort the data frame by the number of points obtained for the exam, from 
highest to lowest. In case of the same number of points, the persons 
should be displayed in alphabetical order.
'''

data = data.sort_values(['name'], ascending=True).sort_values(['points'], ascending=False).reset_index(drop=True)
data
```
