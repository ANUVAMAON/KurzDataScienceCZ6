**Create charts using functions**

```python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color = color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color = color)
  axis.tick_params('y', color = color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['Interpolated'], 'r', 'Time', 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['AverageTemperature'], 'b', 'Time', 'Temperature')
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['AverageTemperature'], 'b', 'Time', 'Temperature')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['Interpolated'], 'r', 'Time', 'CO2')
```

**Histograms**

```python
medals = pd.read_csv("/Users/maro-personal/Documents/KurzDataScience/05_processing_of_data_sets/ML-datasets-master/olympic.csv", index_col='Year')


def groupby_column(dataframe, column):
  groupby = dataframe.groupby(column)
  return({x: groupby.get_group(x) for x in groupby.groups})

all_sports = groupby_column(medals, 'Sport')
```

```python
fig, ax = plt.subplots()
ax.bar("Gymnastics", all_sports["Gymnastics"]['Height'].mean())
ax.bar("Basketball", all_sports["Basketball"]['Height'].mean())
ax.set_ylabel("Height [cm]")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar("Gymnastics", all_sports["Gymnastics"]['Height'].mean())
ax.bar("Basketball", all_sports["Basketball"]['Height'].mean())
ax.set_ylabel("Height [cm]")
ax.set_ylim(0,1000)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(all_sports["Gymnastics"]['Height'], label = "Gymnastics", bins = 20)
ax.hist(all_sports["Basketball"]['Height'], label = "Basketball", bins = 20,
        histtype = "step")
ax.set_xlabel("Height [cm]")
ax.set_ylabel('# of observations')
plt.show()
```
