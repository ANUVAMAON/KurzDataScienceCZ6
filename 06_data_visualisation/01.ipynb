**Create charts using functions**

```python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color = color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color = color)
  axis.tick_params('y', color = color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['Interpolated'], 'r', 'Time', 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['AverageTemperature'], 'b', 'Time', 'Temperature')
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['AverageTemperature'], 'b', 'Time', 'Temperature')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['Interpolated'], 'r', 'Time', 'CO2')
```
